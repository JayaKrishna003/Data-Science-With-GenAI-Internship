```python
# 1.Maximum Product of Two Elements in an Array

def max_product(nums):
    first = second = 0

    for num in nums:
        if num > first:
            second = first
            first = num
        elif num > second:
            second = num

    return (first - 1) * (second - 1)
nums = [3, 4, 5, 2]
print(max_product(nums))
```

```python
# 2.Count Number of Teams

def count_teams(rating):
    teams = 0
    n = len(rating)

    for j in range(n):
        left_smaller = sum(rating[i] < rating[j] for i in range(j))
        left_greater = j - left_smaller

        right_smaller = sum(rating[k] < rating[j] for k in range(j + 1, n))
        right_greater = (n - j - 1) - right_smaller

        teams += left_smaller * right_greater
        teams += left_greater * right_smaller
    return teams
print(count_teams([1, 2, 3, 4]))
```

```python
# 3.Number of Students Doing Homework at a Given Time

def busy_student(startTime, endTime, queryTime):
    count = 0
    for s, e in zip(startTime, endTime):
        if s <= queryTime <= e:
            count += 1
    return count

print(busy_student([1, 2, 3], [3, 2, 7], 4))
```

```python
# 4.Number of Steps to Reduce a Number to Zero

def number_of_steps(num):
    steps = 0
    while num > 0:
        steps += 1
        num = num // 2 if num % 2 == 0 else num - 1
    return steps

print(number_of_steps(14))
```

```python
# 5.Counting Bits

def count_bits(n):
    result = []
    for i in range(n + 1):
        result.append(bin(i).count('1'))
    return result

print(count_bits(5))
```
